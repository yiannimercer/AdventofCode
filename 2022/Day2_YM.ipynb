# Advent of Code \- Day 2

What would your total score be if everything goes exactly according to your strategy guide?


```python
# IMPORT DATA FOR TODAYS PUZZLE
from aocd import data, submit
import pandas as pd
import numpy as np
```

```python
# TAKE A QUICK LOOK AT THE DATA
print(data, end='\n\n')
```

## Part A



### High Level Plan

Establish a rulebook and assign numeric values to each line.  Get round total, and sum all together to get entire game score



### Rulebook

**Opponent Lettering & Scoring**

- A : Rock == 1 Point
- B : Paper == 2 Points
- C : Scissors == 3 Points

**My Lettering**

- X : Rock == 1 Point
- Y : Paper == 2 Points
- Z : Scissors == 3 Points

**Round Scoring** 

- Lost : 0
- Draw : 3
- Won : 6



### Notes & Approach

1. Convert data to pandas data frame
2. Change each letter to it's Rock, Paper, Scissors equivalent
3. Calculate an _Outcome_ column
4. Apply rules to get a _Score_ column
5. Sum _Score_ column



### Code



```python
# SPLIT DATA BY DOUBLE \N
split_data = data.split('\n')
```

```python
# ASSIGN WHAT OPPONENT PLAYS AND I PLAY TO SEPERATE LISTS
opponent_moves = []
my_moves = []

for game in split_data:
    opponent_move = game[0]
    opponent_moves.append(opponent_move)
    my_move = game[2]
    my_moves.append(my_move)
```

```python
# CREATE DF WHERE EACH ROW REPRESENTS EACH ROUND, WITH FIRST COLUMN IS THE OPPONENT'S MOVE AND THE SECOND COLUMN IS MY MOVE
df = pd.DataFrame(list(zip(opponent_moves, my_moves)), columns =['Opponent', 'Me'])

# CHANGE THE LETTERS TO ROCK PAPER SCISSOR MOVES FOR CLARITY, USING THE ABOVE RULE BOOK
df['Opponent_Move'] = np.where(df['Opponent'] == 'A', 'Rock',
                              np.where(df['Opponent'] == 'B', 'Paper',
                                       np.where(df['Opponent'] == 'C', 'Scissors', "REVIEW")))

df['My_Move'] = np.where(df['Me'] == 'X', 'Rock',
                              np.where(df['Me'] == 'Y', 'Paper',
                                       np.where(df['Me'] == 'Z', 'Scissors', "REVIEW")))
```

```python
df.Opponent_Move.value_counts()}
```

```python
df.My_Move.value_counts()
```

```python
# CREATE MOVE SCORE BEFORE GAME OUTCOME FOR MYSELF
df['My_Move_Score'] = np.where(df['My_Move'] == 'Rock', 1,
                                    np.where(df['My_Move'] == 'Paper', 2, 
                                             np.where(df['My_Move'] == 'Scissors', 3, 'REVIEW')))
```

```python
df['My_Move_Score'].value_counts()
```

```python
# CREATE GAME EVALUATOR FUNCTION
def rock_paper_scissors(opp, me):
    if opp == me:
        return 'Draw'
    elif me == 'Rock' and opp == 'Paper':
        return 'Lost'
    elif me == 'Rock' and opp == 'Scissors':
        return 'Won'
    elif me =='Paper' and opp == 'Scissors':
        return 'Lost'
    elif me =='Paper' and opp == 'Rock':
        return 'Won'
    elif me == 'Scissors' and opp == 'Paper':
        return 'Won'
    elif me == 'Scissors' and opp == 'Rock':
        return 'Lost'
    else:
        return "REVIEW"
```

```python
# CREATE OUTCOME FIELD
df['Outcome'] = df.apply(lambda x: rock_paper_scissors(x.Opponent_Move, x.My_Move), axis=1)
```

```python
df['Outcome'].value_counts()
```

```python
# CREATE OUTCOME SCORE FOR EACH ROUND
df['Outcome_Score'] = np.where(df['Outcome'] == 'Lost', 0,
                              np.where(df['Outcome'] == 'Draw', 3,
                                      np.where(df['Outcome'] == 'Won', 6, "REVIEW")))
```

```python
df['Outcome_Score'].value_counts()
```

```python
# CREATE ROUND SCORE COLUMN
df['Round_Score'] = df['My_Move_Score'].astype(int) + df['Outcome_Score'].astype(int)
```

```python
# CALCUALTE TOTAL SCORE
total_score_parta = df['Round_Score'].sum()
print("The total score, if everything goes exactly according to my strategy guide, would be {}".format(total_score_parta))
```

## Submit Answer



```python
submit(total_score_parta)
```

## Part B



### High Level Plan

Establish a rulebook and assign numeric values to each line.  Get round total, and sum all together to get entire game score



### Rulebook

**Opponent Lettering & Scoring**

- A : Rock == 1 Point
- B : Paper == 2 Points
- C : Scissors == 3 Points

**My Lettering**

- X : Lose
- Y : Draw
- Z : Win

**Round Scoring** 

- Lost : 0
- Draw : 3
- Won : 6



### Notes & Approach

1. Convert data to pandas data frame
2. Change each letter to it's Rock, Paper, Scissors, or Lose, Draw, Win equivalent
3. Calculate an _Outcome_ column
4. Apply rules to get a _Score_ column
5. Sum _Score_ column

_We'll need to create a column that is the calculated move column for this part_


### Code



```python
# SPLIT DATA BY DOUBLE \N
split_data = data.split('\n')
```

```python
# ASSIGN WHAT OPPONENT PLAYS AND THE DESIRED OUTCOMES TO SEPERATE LISTS
opponent_moves = []
desired_outcomes = []

for game in split_data:
    opponent_move = game[0]
    opponent_moves.append(opponent_move)
    desired_outcome = game[2]
    desired_outcomes.append(my_move)
```

```python
# CREATE DF WHERE EACH ROW REPRESENTS EACH ROUND, WITH ONE COLUMN REPRESENTING THE DESIRED OUTCOME AND ANOTHER COLUMN IS THE OPPONENTS MOVE
df = pd.DataFrame(list(zip(opponent_moves, my_moves)), columns =['Opponent', 'Desired_Outcome'])

# CHANGE THE LETTERS TO ROCK PAPER SCISSOR MOVES FOR CLARITY, USING THE ABOVE RULE BOOK
df['Opponent_Move'] = np.where(df['Opponent'] == 'A', 'Rock',
                              np.where(df['Opponent'] == 'B', 'Paper',
                                       np.where(df['Opponent'] == 'C', 'Scissors', "REVIEW")))

df['Desired_Outcome'] = np.where(df['Desired_Outcome'] == 'X', 'Lose',
                              np.where(df['Desired_Outcome'] == 'Y', 'Draw',
                                       np.where(df['Desired_Outcome'] == 'Z', 'Win', "REVIEW")))
```

```python
df.Opponent_Move.value_counts()
```

```python
df.Desired_Outcome.value_counts()
```

```python
# CREATE CALCULATE MOVE FUNCTION
def move_calculator(desired_outcome, opp_move):
    if desired_outcome == 'Draw':
        return opp_move
    elif desired_outcome == 'Lose' and opp_move == 'Paper':
        return "Rock"
    elif desired_outcome == 'Lose' and opp_move == 'Scissors':
        return "Paper"
    elif desired_outcome == 'Lose' and opp_move == 'Rock':
        return "Scissors"
    elif desired_outcome == 'Win' and opp_move == 'Paper':
        return "Scissors"
    elif desired_outcome == 'Win' and opp_move == 'Scissors':
        return "Rock"
    elif desired_outcome == 'Win' and opp_move == 'Rock':
        return "Paper"
    else:
        return "REVIEW"
```

```python
# CREATE CALCULATED MOVE COLUMN
df['My_Move'] = df.apply(lambda x: move_calculator(x.Desired_Outcome, x.Opponent_Move), axis=1)
```

```python
# CREATE MOVE SCORE BEFORE GAME OUTCOME FOR MYSELF
df['My_Move_Score'] = np.where(df['My_Move'] == 'Rock', 1,
                                    np.where(df['My_Move'] == 'Paper', 2, 
                                             np.where(df['My_Move'] == 'Scissors', 3, 'REVIEW')))
```

```python
df['My_Move_Score'].value_counts()
```

```python
# CREATE OUTCOME SCORE FOR EACH ROUND
df['Outcome_Score'] = np.where(df['Desired_Outcome'] == 'Lose', 0,
                              np.where(df['Desired_Outcome'] == 'Draw', 3,
                                      np.where(df['Desired_Outcome'] == 'Win', 6, "REVIEW")))
```

```python
df['Outcome_Score'].value_counts()
```

```python
# CREATE ROUND SCORE COLUMN
df['Round_Score'] = df['My_Move_Score'].astype(int) + df['Outcome_Score'].astype(int)
```

```python
# CALCUALTE TOTAL SCORE
total_score_partb = df['Round_Score'].sum()
print("The total score, if everything goes exactly according to my NEW UPDATED strategy guide, would be {}".format(total_score_partb))
```

```python
submit(total_score_partb)
```
