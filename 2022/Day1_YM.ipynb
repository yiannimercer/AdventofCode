## Part A

### High Level Plan

Get each elves' food into separate lists, sum each list, and filter to the highest value

### Notes & Approach

We know that each elf's food inventory is separated by a new line \(\\n\)

1. Split the data on each blank 
2. Create high level list container to store each elves food inventory
3. Loop through the split data and append each food item into a sublist.  If the value is a blank string, create a new list and add it to the high level list container
4. Sum each list contained in the high level list container and print the maximum value



### Code

```python
# SPLIT INPUT DATA ON A NEW LINE
split_data = data.split('\n')

# CREATE LIST CONTAINER TO HOLD THE SUB-LISTS, WHICH REPRESENT EACH INDIVIDUAL ELVES' FOOD INVENTORY
all_elvels_food = [[]]

# LOOP THROUGH SPLIT DATA -- IF A VALUE/STRING IS '', ADD A BLANK NEW LIST TO THE ALL EVLES FOOD INVENTORY LIST | THE VALUE/STRING IS IN FACT A FOOD ITEM, ADD IT IN THE LAST OF THE LIST OF LISTS
for food_item in split_data:
    if food_item == '':
        all_elvels_food.append([])
    else:
        all_elvels_food[-1].append(int(food_item)) # CONVERT TO FLOAT WHEN APPENDING (MAKES NEXT CELL IS EASIER)
```

```python
# SUM EACH SUB LIST (WHICH REPRESENTS EACH ELVES' FOOD INVENTORY) & PRINT/SAVE THE HIGHEST VALUE 

## LIST TO SAVE SUMS
total_calories_by_elf = []

## LOOP THROUGH EACH LIST/FOOD INVENTORY AND SUM THE CALORIES --> SAVE THE SUM TO A NEW LIST
for elves_food_inv in all_elvels_food:
    total_calories_by_elf.append(sum(elves_food_inv))

## PRINT/SAVE THE MAXIMUM/HIGHEST CALORIE COUNT
answer_1a = int(max(total_calories_by_elf))
print("The most calories carried by an elf is {} kcals".format(answer_1a))
```

### Submit Part A



```python
submit(answer_1a, part="a", day=1, year=2022)
```

## Part B

### High Level Plan

Take the above list of calorie sums and filter to the top 3 highest values

### Notes & Approach

Just be mindful of how you're submitting the data (i.e., what data type)

### Code

```python
# USE NLARGEST FUNCTION FROM HEAPQ LIBRARY
answer_1b = sum(nlargest(3, total_calories_by_elf))
```

```python
# PRINT THE SUM OF THE THREE LARGEST VALUES/TOTAL CALORIES
print("The sum of the highest three calories totals for all elves is {} kcals".format(answer_1b))
```

### Submit Part B



```python
submit(answer_1b, part='b', day=1, year=2022)
```
